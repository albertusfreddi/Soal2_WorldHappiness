```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2019.csv')
df
```

#### dari feature Score

```python
scoreMean = df['Score'].describe()['mean']
scoreMedian = df['Score'].describe()['50%']
scoreModus = df['Score'].mode()[0]
scoreRange = df['Score'].max() - df['Score'].min()
scoreQ1 = df['Score'].describe()['25%']
scoreQ3 = df['Score'].describe()['75%']
scoreIQR = scoreQ3 - scoreQ1
scoreStd = df['Score'].describe()['std']
scoreVar = df['Score'].var()
scoreZ = df[['Score']].apply(zscore)
scoreHighOutliersIQR = scoreQ3 + (1.5 * scoreIQR)
scoreLowOutliersIQR = scoreMean - (2 * scoreIQR)
scoreHighOutliersZ = scoreMean + (2 * scoreStd)
scoreLowOutliersZ = scoreMean - (2 * scoreStd)

print('Mean dari feature Score adalah : ', scoreMean)
print('Median dari feature Score adalah : ', scoreMedian)
print('Modus dari feature Score adalah : ', scoreModus)
print('Range dari feature Score adalah : ', scoreRange)
print('Q1 dari feature Score adalah : ', scoreQ1)
print('Q3 dari feature Score adalah : ', scoreQ3)
print('IQR dari feature Score adalah : ', scoreIQR)
print('Standard Deviasi dari feature Score adalah : ', scoreStd)
print('Variance dari feature Score adalah : ', scoreVar)
print('High Outliers dengan metode IQR dari feature Score adalah : ', scoreHighOutliersIQR)
print('Low Outliers dengan metode IQR dari feature Score adalah : ', scoreLowOutliersIQR)
print('High Outliers dengan metode Z-score dari feature Score adalah : ', scoreHighOutliersZ)
print('Low Outliers dengan metode Z-score dari feature Score adalah : ', scoreLowOutliersZ)
print('Z-score untuk setiap point dari feature Score adalah : \n', scoreZ)

plt.figure(figsize = (20, 10))

plt.subplot(121)
plt.title('Box plot for Feature Score')
plt.boxplot(df['Score'])

plt.subplot(122)
plt.title('Histogram plot for Feature Score')
lbins = (df['Score'].max() - df['Score'].min()) / 10 # 100 adalah total bar yang di mau
min = df['Score'].min() - (0.5*df['Score'].min())
max = df['Score'].max() + (0.5*df['Score'].min())
bins = np.arange(min, max, lbins)
plt.hist(df['Score'], bins)
plt.show()
```

#### Menentukan
- Mean, Median dan Modus
- Range, Q1, Q3 dan IQR
- Standard Deviasi & Variance
- Z-score tiap data point
- IQR method
- Z-score method
